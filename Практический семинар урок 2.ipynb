Для просмотра статистических данных с типом 'object'
Параметр include = 'object' указывает, что нужна информация только о столбцах с типом данных object (текстовые данные).
Таким образом, данный код вернет следующую сводную информацию о текстовых столбцах DataFrame:
количество уникальных значений(unique);
наиболее часто встречающееся значение (top);
количество встречаемости наиболее часто встречающегося значения (freq).

```python
df.describe(include = 'object')
```
